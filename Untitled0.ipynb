```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'photo1720771445.jpeg'  # Ganti dengan jalur gambar Anda
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Display the original image
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
```

```python
# Apply median filtering using OpenCV
median_filtered = cv2.medianBlur(image, 5)
```

```python
# Display the median filtered image
plt.subplot(1, 3, 2)
plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filtering')
plt.axis('off')
```

```python
# Manual mean filtering
def manual_mean_filtering(image, kernel_size=5):
    # Get image dimensions
    rows, cols = image.shape
    # Define the output image
    output_image = np.zeros_like(image)
    # Define the padding size
    pad_size = kernel_size // 2
    # Pad the original image with zeros
    padded_image = np.pad(image, pad_size, mode='constant', constant_values=0)
    # Iterate over every pixel of the image
    for i in range(rows):
        for j in range(cols):
            # Extract the neighborhood
            neighborhood = padded_image[i:i+kernel_size, j:j+kernel_size]
            # Compute the mean of the neighborhood
            mean_value = np.mean(neighborhood)
            # Assign the mean value to the output image
            output_image[i, j] = mean_value

    return output_image
```

```python
# Apply manual mean filtering
mean_filtered = manual_mean_filtering(image, kernel_size=5)
```

```python
# Display the mean filtered image
plt.subplot(1, 3, 3)
plt.imshow(mean_filtered, cmap='gray')
plt.title('Mean Filtering')
plt.axis('off')

# Show all the images
plt.show()
```
